```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("student_dataset.csv") 
print(df.head(10))
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

drop unnamed column as of no use 

```python
df=df.drop("Unnamed: 0", axis=1) 
# axis=1 i.e vertical alignment
print(df.head(5))
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender",palette = "Set1")
ax.bar_label(ax.containers[0])
plt.title("Gender distrbution",fontsize=12)
plt.show()
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,5))
plt.title("Relationship between Parent Education and Student Scores")
sns.heatmap(gb,annot=True)
plt.show()
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
plt.title("Relationship between Parent Marital Status and Student Scores ")
sns.heatmap(gb1,annot=True)
plt.show()
```
